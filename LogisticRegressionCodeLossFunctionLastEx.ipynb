```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def loss(y, y_pred):
  return -(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))

y = 0
y_pred = 0.97
print(f'Loss for y={y}, y_pred={y_pred}: {loss(y, y_pred):.4f}')
```

```python
y_pred_values = np.linspace(0.01, 0.99, 100)

plt.plot(y_pred_values, loss(1, y_pred_values), label='y = 1') # loss function for y = 1

plt.plot(y_pred_values, loss(0, y_pred_values), label='y = 0')

plt.xlabel('Predicted Probability (y_pred)')
plt.ylabel('Loss')
plt.title("Loss function for Logistic Regression")
plt.legend()
plt.grid(True)
```
